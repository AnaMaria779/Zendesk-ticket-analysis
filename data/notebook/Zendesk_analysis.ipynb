```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\anazu\\OneDrive\\Escritorio\\zendesk_tickets.csv")
df.head()
df.info()
df.describe()
```

```python
issue_counts = df["Issue Type"].value_counts()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=issue_counts.index, y=issue_counts.values, hue=issue_counts.index, legend=False, palette="viridis")
plt.title("Most Common Customer Issues")
plt.xticks(rotation=45)
plt.show()
```

```python
avg_response_time = df["Response Time (hours)"].mean()
avg_resolution_time = df["Resolution Time (hours)"].mean()
print(f"Average Response Time: {avg_response_time:.2f} hours")
print(f"Average Resolution Time: {avg_resolution_time:.2f} hours")
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Response Time (hours)"], bins=10, kde=True, color="blue")
plt.title("Distribution of Response Times")
plt.xlabel("Response Time (hours)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Resolution Time (hours)"], color="green")
plt.title("Resolution Time Distribution")
plt.xlabel("Resolution Time (hours)")
plt.show()
```
